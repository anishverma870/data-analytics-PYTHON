```python
#This process does not remove outliers but caps them, meaning it brings extreme values closer to the normal range, which can improve model performance and reduce skew.
def cap_outliers(df, columns):
    df_capped = df.copy()

    for column in columns:
        Q1 = df_capped[column].quantile(0.25)
        Q3 = df_capped[column].quantile(0.75)
        IQR = Q3 - Q1

        if IQR == 0:
            continue  # Skip capping if there's no variation

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df_capped[column] = df_capped[column].apply(lambda x:
            lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))

    return df_capped
```

```python
# used the cap_outliers function on the columns that are likely to contain outliers:
columns_to_treat = ['OBS_30_CNT_SOCIAL_CIRCLE', 'DEF_30_CNT_SOCIAL_CIRCLE', 'OBS_60_CNT_SOCIAL_CIRCLE','DEF_60_CNT_SOCIAL_CIRCLE']
df_cleaned = cap_outliers(inp0, columns_to_treat)
```
