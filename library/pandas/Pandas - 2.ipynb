```python
tips.drop('cust_seg',axis=1,inplace=True)
```

```python
def cust_seg(rahul):
    if rahul>250:
        return 'Platinum'
    elif rahul>200 and rahul<=250:
        return 'Gold'
    elif rahul>150 and rahul<=200:
        return 'Silver'
    else:
        return 'Bronze'
```

```python
tips
```

```python
tips['cust_seg'] = tips['total_rev'].map(cust_seg)
```
