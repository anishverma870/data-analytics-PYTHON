# Line Plot

### Example
Line plot

Plotting the stock prices of Amazon varied over the years as per the data in the AMZN stocks dataset

```python
# Opening price of Amazon stock over time
plt.figure(figsize = (6, 4))
sns.lineplot(data = amzn, x = 'Date', y = 'Open')
plt.xlabel('Date')
plt.ylabel('Opening price in dollars')
plt.title('Opening price of Amazon stock')
plt.xticks(range(0, 3019, 302), rotation = 45);
```

Plotting both opening and closing prices and adding a legend -

```python
# Price of Amazon stock over time
plt.figure(figsize = (6, 4))
sns.lineplot(data = amzn, x = 'Date', y = 'Open', label = 'Opening price')
sns.lineplot(data = amzn, x = 'Date', y = 'Close', label = 'Closing price')
plt.xlabel('Date')
plt.ylabel('Price in dollars')
plt.title('Price of Amazon stock')
plt.xticks(range(0, 3019, 302), rotation = 45)
plt.legend();
```

Plotting the stock prices for 100 days -

```python
sub_amzn = amzn[0:100]
```

```python
# Price of Amazon stock over 100 days
plt.figure(figsize = (6, 4))
sns.lineplot(data = sub_amzn, x = 'Date', y = 'Open', label = 'Open price')
sns.lineplot(data = sub_amzn, x = 'Date', y = 'Close', label = 'Close price')
plt.xlabel('Date')
plt.ylabel('Price in dollars')
plt.title('Price of Amazon stock')
plt.xticks(range(0, 100, 10), rotation = 45)
plt.legend();
```

# Bar Graphs

### Example

Plotting customers opting for each payment method -

```python
# Number of customers for each payment method
plt.figure(figsize = (6, 4))
sns.countplot(data = df, x = 'PaymentMethod')
plt.title('Number of customers per payment method')
plt.xlabel('Payment method')
plt.ylabel('Count');
```

Changing the vertical bar graph to a horizontal one -

```python
# Number of customers for each payment method
plt.figure(figsize = (6, 4))
sns.countplot(data = df, y = 'PaymentMethod', order = df['PaymentMethod'].value_counts().index)
plt.title('Number of customers by payment method')
plt.ylabel('Payment method')
plt.xlabel('Count');
```

Plotting horizontal grouped bar charts -

```python
# Number of customers for each payment method split over billing type
plt.figure(figsize = (6, 4))
sns.countplot(data = df, y = 'PaymentMethod', hue = 'PaperlessBilling', order = df['PaymentMethod'].value_counts().index)
plt.title('Number of customers by payment method split over billing type')
plt.xlabel('Count')
plt.ylabel('Payment method')
plt.legend(title = 'Paperless billing');
```

### Example

Plotting the average charges charges by payment method -

```python
# Average charges per payment method
plt.figure(figsize = (6, 4))
sns.barplot(data = df, y = 'PaymentMethod', x = 'TotalCharges', estimator = 'mean', errorbar = None)
plt.xlabel('Mean total charges')
plt.ylabel('Payment method')
plt.title('Average charges per payment method');
```

Plotting the median charges by payment method and adding an error bar (you can read more about error bars [here](https://seaborn.pydata.org/tutorial/error_bars.html)) -

```python
# Median charges per payment method
plt.figure(figsize = (6, 4))
sns.barplot(data = df, y = 'PaymentMethod', x = 'TotalCharges', estimator = 'median', errorbar = 'sd')
plt.xlabel('Median total charges')
plt.ylabel('Payment method')
plt.title('Median charges per payment method');
```

Plotting a grouped bar chart for average charges by payment method -

```python
# Average charges per payment method split by billing type
plt.figure(figsize = (6, 4))
bar_plot = sns.barplot(data = df, y = 'PaymentMethod', x = 'TotalCharges', hue = 'PaperlessBilling', estimator = 'mean', errorbar = None)
bar_plot.set_xlabel('Mean total charges')
bar_plot.set_ylabel('Payment method')
bar_plot.set_title('Average charges per payment method split by billing type');
```

Using a facet grid to plot bar graphs for average charges by payment method -

```python
# Average total charges per payment method split by billing type
bar_plot = sns.catplot(data = df, kind = 'bar', y = 'PaymentMethod', x = 'TotalCharges', col = 'PaperlessBilling', estimator = 'mean', errorbar = None)
bar_plot.set_axis_labels('Mean total charges', 'Payment method')
bar_plot.fig.suptitle('Average total charges per payment method split by billing type', y = 1.02);
```

The syntax is different here because we are editing the facet grid. You can learn more about it [here](https://seaborn.pydata.org/generated/seaborn.FacetGrid.html).

# Plotting quantiles
In pandas, **quantiles** can be computed using the `quantile()` method, which provides a way to divide a dataset into intervals based on the distribution of the data. Quantiles are points in a dataset that correspond to certain probabilities. The `describe()` method gives you quartiles by default.

### Example

Plotting the distribution of tenure of customers in the churn dataset

# Heat maps
Heat maps are visualisations that map numerical scales to color scales. This helps in visualising areas of larger numerical values and smaller numerical values. One of the most common metric that is visualised using a heat map is the correlation coefficient.

### Example
**Correlation**: It is a statistical measure that quantifies the strength and direction of a linear relationship between two variables. It is commonly expressed as the correlation coefficient, which ranges from $-1$ to $1$. Values closer to $0$ indicate weak relationships while values farther away from $0$ indicate stronger relationships. A negative value indicates an inverse relationship while a positive value indicates a direct relationship.

```python
df_num_sub = df[['tenure', 'MonthlyCharges', 'TotalCharges']]
df_num_sub.head()
```

```python
df_num_sub = df_num_sub.convert_dtypes()
df_num_sub.info()
```

```python
# Create correlation matrix for provided dataframe
corr_mat = df_num_sub.corr()
corr_mat
```

```python
# Rounding off to two digits
corr_mat.round(2)
```

### Example

Plotting correlation between numerical variables in the churn dataset

```python
# Heatmap showing correlation between tenure, monthly charges, and total charges
plt.figure(figsize = (10, 6))
sns.heatmap(data = corr_mat)
plt.title('Heatmap showing correlation between tenure, monthly charges, and total charges');
```

Customising the heatmap - added labels, changed colour, changed line width

```python
# Heatmap showing correlation between tenure, monthly charges, and total charges
plt.figure(figsize = (10, 6))
sns.heatmap(data = corr_mat, annot = True, cmap = 'coolwarm', fmt = '.2f', linewidths = 0.5)
plt.title('Heatmap showing correlation between tenure, monthly charges, and total charges');
```

# Pair plot
A pair plot is a collection of plots arranged in a tabular fashion. Each plot is a visualisation between two numerical variables, one from the X-axis and one from the Y-axis. If both variables are different, you will get a scatter plot (the off-diagonal elements). If they are the same, you will get a histogram (the diagonal elements). Pair plots are used to quickly analyse numerical variables in datasets.

Also, note the following:
1. x-axis in histograms: Shows the variables values (same as in scatter plots).
2. y-axis in histograms: Represents the variable's value too, not the frequency or density. The height of the bars is not "true" frequency but rather adjusted for consistency with the other plots.

### Example

Studying the numerical features Open, Close and Volume in the `amzn` dataset

```python
# Pairplot of Open, Close and Volume
sns.pairplot(amzn[['Open', 'Close', 'Volume']])
```

# Kernel density estimation (KDE) plots
The KDE plot is a smooth and continuous estimate of a histogram of a numerical variable. The Y-axis represents the probability density instead of the frequency of the X-value.

### Example
KDE plots can complement histograms by providing a continuous overview of the distribution, while histograms offer a discrete representation.

Plotting the distribution of tenure in the dataset - using subplots

```python
# Distribution of tenure of customers
plt.figure(figsize = (8, 4))
plt.subplot(1, 2, 1)
sns.histplot(data = df, x = 'tenure')
plt.title('Distribution of tenure of customers', fontsize = 12)
plt.xlabel('Tenure', fontsize = 10)
plt.ylabel('Frequency', fontsize = 10)
plt.subplot(1, 2, 2)
sns.kdeplot(data = df, x = 'tenure', fill = True)
plt.title('Distribution of tenure of customers', fontsize = 12)
plt.xlabel('Tenure', fontsize = 10)
plt.ylabel('Probability density', fontsize = 10)
plt.tight_layout();
```

Plotting the distribution of tenure in the dataset - adding a KDE plot on top of a histogram

```python
# Distribution of tenure of custormers
plt.figure(figsize = (8, 4))
sns.histplot(data = df, x = 'tenure', kde = True)
plt.title('Distribution of tenure of customers', fontsize = 12)
plt.xlabel('Tenure', fontsize = 10)
plt.ylabel('Frequency', fontsize = 10)
plt.tight_layout();
```
