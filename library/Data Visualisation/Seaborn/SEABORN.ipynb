# SEABORN

## Plot Creation using Seaborn

It is an extension of Matplotlib.

It comes with various interesting default themes and visualisation patterns; hence, it can be used to summarise the visualisation of an entire/complex data set.

It avoids overlapping of plots, i.e., customisation is not required in most cases, as default themes serve the purpose.

It treats the whole data set as a single unit. Parameters are required while calling methods such as plot().

As it is built upon Matplotlib, it uses Matplotlib along with NumPy and Pandas for plotting graphs.

Simple tasks are much easier with Seaborn, for example, adding an extra argument name ‘hue’ while plotting a graph adds colour.

# Bar graph: 
Helpful when you need to visualise a numeric feature (fact) across multiple categories.

```python
# important library
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data = {
    'c':10,
    'c+':20,
    'java':15,
    'python':40
}
course = list(data.keys())
values = list(data.values())
# making data fram for the seaborn 
df = pd.DataFrame({'course':course,'values':values})
#making color pallete for seaborn
color = sb.color_palette(['red','blue','green','yellow'])
# Plot the bar chart with color coding
ax = sb.barplot(x='course',hue='course',y = 'values',data = df,palette=color)# hue, Seaborn assigns different colors to each unique value in that column.
plt.show()
```

### Customisation

```python
import pandas as pd 
import seaborn as sb
```

```python
data2 = {
    'c':10,
    'c+':15,
    'java':13,
    'python':40
}
course2 = list(data2.keys())
values2 = list(data2.values())
df2 = pd.DataFrame({'course2':course2,'values2':values2})
color2 = sb.color_palette(['orange','pink','brown','black'])
sb.set_style('whitegrid',{'grid.color':'red','grid.linestyle':'--'})
ax2 = sb.barplot(x='course2',y = 'values2',hue='course2',data=df2,palette=color2)
```

# Scatter plot: 
Used to identify a relationship or pattern between two quantitative numeric variables and the present of outliers within them.

```python
x =[5, 7, 8, 7, 2, 17, 2, 9,

    4, 11, 12, 9, 6]

 

y =[99, 86, 87, 88, 100, 86,

    103, 87, 94, 78, 77, 85, 86]
```

```python
sb.scatterplot(x=x,y=y)
sb.set_style('whitegrid',{'grid.color':'blue','grid.linestyle':'--'})
```

# Line chart: 
Depicts the trend of a variable over a specified period; works with both numerical and categorical data.

```python
ax = sb.lineplot(x=x,y=y,legend=True)
plt.title('LINE CHART',color = 'red',size = 40)
plt.show()
```

# Histogram:
It is used to represent data in the form of groups. It sets ranges for data (bins) and then distributes the data across these bins.

```python
data = np.random.randint(100, 150, 100)  # Generates 10 random integers from 1 to 99
print(data)
```

## The kde 
parameter stands for kernel density estimate and is used to smooth the distribution and draw a density plot line.

```python
sb.histplot(data = data,color = 'brown',kde=True,bins=20)
plt.title('HISTOGRAM',color = 'blue',size = 45)
plt.xlabel('DATA')
plt.show()
```

# Distplot: 
It depicts how a single variable is distributed across a data set. It combines a histogram and a kde in a graph.

```python
sb.displot(data = data,kde = True,color ='r')
plt.title('DISPLOT',color = 'black',size = 45)
plt.show()
```

```python
sb.displot(data = data,kind='kde',color ='r')
plt.show()
```

# Boxplot: 
It is used to see the quartile-wise distribution of any continuous variable against a categorical variable.

```python
 ax = sb.boxplot(data)
```
